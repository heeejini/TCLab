```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 경로 (사용자 파일 시스템에 맞게 수정 필요)
csv_path = "rollout.csv"  # 경로 수정 가능

# CSV 파일 로드
df = pd.read_csv(csv_path)

# 그래프 그리기
fig, axs = plt.subplots(2, 1, figsize=(14, 8), sharex=True)

# 온도 추적 시각화
axs[0].plot(df["time"], df["T1"], label="T1 (Measured)", linewidth=2)
axs[0].plot(df["time"], df["TSP1"], label="TSP1 (Setpoint)", linestyle="--")
axs[0].plot(df["time"], df["T2"], label="T2 (Measured)", linewidth=2)
axs[0].plot(df["time"], df["TSP2"], label="TSP2 (Setpoint)", linestyle=":")
axs[0].set_ylabel("Temperature (°C)")
axs[0].set_title("Temperature Tracking")
axs[0].grid(True)
axs[0].legend()

# 히터 출력 시각화
axs[1].plot(df["time"], df["Q1"], label="Q1 (Heater 1 Output)", linewidth=2)
axs[1].plot(df["time"], df["Q2"], label="Q2 (Heater 2 Output)", linewidth=2)
axs[1].set_ylabel("Heater Output (%)")
axs[1].set_xlabel("Time (s)")
axs[1].set_title("Heater Output Over Time")
axs[1].grid(True)
axs[1].legend()

plt.tight_layout()
plt.show()
```
